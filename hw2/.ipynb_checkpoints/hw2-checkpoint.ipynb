## Question 1

*Problem 4.11(a,c,e) of Boyd and Vandenberghe*


Problems involving $\ell1$ and $\ell \infty$ norms. Formulate the following problems as LPs. Explainin detail the relation between the optimal solution of each problem and the solution of its equivalent LP

In each problem, $A \in R^{m√ón}, x \in R^{n}, b \in R^{m}$

### Question 1a

$\text{minimize }\vert\vert Ax-b \vert\vert_{\infty}$

### Solution

$\text{let } \mathbf{1} \in R^{m \times 1}, t \in R$

$\text{minimize }\vert\vert Ax-b \vert\vert_{\infty}$

$\begin{align*}
\equiv \text{minimize } &t \\
\text{s.t. }& -t \cdot \mathbf{1} \le Ax-b \le t \cdot \mathbf{1} \\
&t \ge 0
\end{align*}$


### Question 1b

$\text{minimize }\vert\vert Ax-b \vert\vert_{1}$

### Solution

$\text{let } t \in R^m_{+}, \mathbf{1} \in R^m$

$\text{minimize } \vert\vert Ax-b \vert\vert_{1}$

$\begin{align*}
\equiv \text{minimize } & \sum t \\
\text{s.t. }& -t \le Ax-b \le t\\
\end{align*}$

$\begin{align*}
\equiv \text{minimize } & \mathbf{1}^T t \\
\text{s.t. }& -t \le Ax-b \le t\\
\end{align*}$


### Question 1c

$\text{minimize }\vert\vert Ax-b \vert\vert_{1}$

$\text{subject to } \vert\vert x \vert\vert_{\infty} \le 1$

### Solution

$\text{let } t \in R^m_{+}, \mathbf{1} \in R^m$

$\text{minimize }\vert\vert Ax-b \vert\vert_{1}$

$\begin{align*}
\equiv \text{minimize } & \mathbf{1}^T t \\
\text{s.t. }& -t \le Ax-b \le t\\
& -\mathbf{1}\le x \le \mathbf{1}
\end{align*}$


### Question 1d

$\text{minimize } \vert\vert x \vert\vert_{1}$

$\text{subject to }\vert\vert Ax-b \vert\vert_{\infty} \le 1$

### Solution
$\text{Let } \mathbf{1} \in R^n, t \in R^n$

$\text{minimize } \mathbf{1}^T t$

$\begin{align*} 
\text{subject to } -\mathbf{1} \le &Ax-b \le \mathbf{1}\\
&-t\le x \le t
\end{align*}$




### Question 1e

$\text{minimize }\vert\vert Ax-b \vert\vert_{1} + \vert\vert x \vert\vert_{\infty}$

### Solution
$\text{Let } \mathbf{1} \in R^m, y \in R^n, t \in R^m$

$\text{minimize } \mathbf{1}^T t + y$

$\begin{align*} 
\text{subject to } &-t \mathbf{1}  \le Ax-b \le t\mathbf{1}\\
& -y \le x \le y\\
\end{align*}$



## Question 2

*Minimum fuel optimal control*. We consider a linear dynamical system with state $x(t) \in R^n , t = 0, . . . , N$ , and actuator or input signal $u(t) \in R$, for $t = 0, . . . , N - 1$. The dynamics of the system is given by the linear recurrence


$x(t + 1) = Ax(t) + bu(t), t = 0, . . . , N - 1$,

where $A \in R^{n \times n}$ and $b \in R^n$ are given. We assume that the initial state is zero, i.e., $x(0) = 0$


The minimum fuel optimal control problem is to choose the inputs $u(0), . . . , u(N - 1)$ so as to minimize the total fuel consumed, which is given by

$\displaystyle F = \sum_{t=0}^{N - 1} f (u(t)), t=0,.....,N-1$

subject to the constraint that $x(N ) = x_{des}$ , where N is the (given) time horizon, and $x_{des} \in R^n$ is the (given) desired final or target state. The function $f : R \rightarrow R$ is the fuel use map for the actuator, and gives the amount of fuel used as a function of the actuator signal amplitude. In this problem we use


$f(a) = \begin{cases}
|a| &  |a|\leq 1 \\
2|a|-1 & |a|\gt 1
\end{cases}$



This means that fuel use is proportional to the absolute value of the actuator signal, for actuator signals between -1 and 1; for larger actuator signals the marginal fuel efficiency is half.

Formulate the minimum fuel optimal control problem as an LP.



### Solution (To be a simplified version of the problem)

#### Note:

---
Simplify the problem a ltitle. As not sure how to deal with the cases in the original problem

$\text{Let } f(a) = |a|$

-----


$\begin{align*}
\text{Minimize } &\displaystyle \sum_{t=0}^{N - 1} f (u(t)), t=0,.....,N-1\\
\text{Subject to } & x(N) = x_{des}\\
\end{align*}$

Equivalent to:

Let variables, $f(0)...f(N-1) \in R, u(0) ... u(N-1) \in R, x(0)...x(N) \in R^n$

$\begin{align*}
\text{Minimize } & f(0) + f(1) .... f(N-1)\\
\text{Subject to } & f(t) \ge 0, t=0,...,N-1 \\
& -f(t) \le u(t) \le f(t), t=0,....,N-1\\
&x(0) = 0 \\
&x(t + 1) = Ax(t) + bu(t), t = 0,.....,N-1\\
& x(N) = x_{des}\\
\end{align*}$




Using N=2 as an example, the linear program is:


$\begin{align*}
\text{Minimize } & f(0) + f(1)\\
\text{Subject to } & f(0) \ge 0\\ 
& f(1) \ge 0\\
& -f(0) \le u(0) \le f(0)\\ 
& -f(1) \le u(1) \le f(1)\\ 
& x(0) = 0 \\
&x(1) = Ax(0) + bu(0)\\
&x(2) = Ax(1) + bu(1)\\
& x(2) = x_{des}\\
\end{align*}$


### Solution (To the original problem)

```python
# piecewise-linear optimization
# http://www.seas.ucla.edu/~vandenbe/ee236a/lectures/pwl.pdf
x = np.linspace(-2, 2, 50)
y1 = np.piecewise(x, [np.abs(x) <= 1, np.abs(x) >= 1], [lambda x: np.abs(x), lambda x: 2*np.abs(x)-1])
y_prime = np.piecewise(x, [np.abs(x) >= 1, np.abs(x) <= 1], [lambda x: np.abs(x), lambda x: 2*np.abs(x)-1])

fig = plt.figure(figsize=(10,15))
ax = fig.add_subplot(311)
ax.spines['left'].set_position('center')
ax.plot(x, y1)
ax.plot(x, y_prime, 'r--')
ax.set_ylim(0,3)
ax.set_title("Original Equation")

y2 = np.maximum(np.abs(x), 2*np.abs(x)-1)

ax = fig.add_subplot(312)
ax.spines['left'].set_position('center')
ax.set_title("Transformed Equation")
ax.text(1,1, r'$max(|a|, 2|a|-1)$')
ax.plot(x, y2)


ax = fig.add_subplot(313)
ax.spines['left'].set_position('center')
ax.set_title("Both")
ax.plot(x, y2, x, y1)
```

& x(N) = x_{des}\\
\end{align*}$
