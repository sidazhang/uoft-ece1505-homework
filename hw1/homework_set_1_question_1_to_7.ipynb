## Problem 4c

Suppose that $f:R^n\times R^m \to R$ is differentiable, but not necessarily convex-concave, and the saddle-point property holds at $\tilde{x},\tilde{z}$

$f(\tilde{x}, z)\leq f(\tilde{x}, \tilde{z}) \leq f(x, \tilde{z})$, $\forall x, z$

Show that $\nabla f(\tilde{x}, \tilde{z}) = 0$

## Solution

$\nabla f(x,z) = \left[\begin{smallmatrix} \frac{\partial{f}}{\partial{x}} \\ \frac{\partial{f}}{\partial{z}}\end{smallmatrix}\right]$

$\because f(\tilde{x}, \tilde{z}) \leq f(x, \tilde{z})\forall x$

$\therefore f(\tilde{x}, \tilde{z})$ is the minimum point at fixed $z$
$\implies \frac{\partial{f}}{\partial{x}} = 0$ at $(\tilde{x}, \tilde{z})$

Also

$\because f(\tilde{x}, z)\leq f(\tilde{x}, \tilde{z})\forall z$

$\therefore f(\tilde{x}, \tilde{z})$ is the max point at fixed $x$
$\implies \frac{\partial{f}}{\partial{z}} = 0$ at $(\tilde{x}, \tilde{z})$

So

$\nabla f(\tilde{x}, \tilde{z}) = \left[\begin{smallmatrix} \frac{\partial{f}}{\partial{x}} \\ \frac{\partial{f}}{\partial{z}}\end{smallmatrix}\right] = \left[\begin{smallmatrix} 0 \\0\end{smallmatrix}\right] = 0$

$f$ is convex if $\nabla^2 f$ is P.S.D

Let $\nabla^2 f = H$

$H = \left[\begin{smallmatrix}2 & \beta\\\beta & 2\end{smallmatrix}\right]$


### At what values of $\beta$ is H is P.S.D

**Approach 1: $R^{2\times 2} trick$**

With a 2 by 2 matrix

$H=\begin{bmatrix} a & b\\ c & d\end{bmatrix}$

Matrix is P.S.D
if $a \geq 0$ and $det \geq 0$

In this case, $a = 2 \geq 0$ and $det = 4 - \beta^2$ 

So $det \geq = 0$ when $-2 \leq \beta \leq 2$

```python
## approach 2 solution
v1, v2 = symbols("v_1,v_2")
v = Matrix([v1, v2])
h_equation = simplify((v.T * hessian * v)[0])

L_h_equation = get_latex(h_equation)
```

**Approach 2: H is P.S.D when $v^tHv \geq 0$**

$v^tHv=2 \beta v_{1} v_{2} + 2 v_{1}^{2} + 2 v_{2}^{2}$

Complete the square on $v_1$ gives us:

$ v_{1}^2 + \beta v_1v_2 + v_2^2 = (v_1+\beta v_2/2)^2 +\frac{4-\beta^2}{4}v_2^2 $

The whole equation is greater than 0 if $\frac{4-\beta^2}{4} \geq 0$

Which gives us

$-2 \leq \beta \leq 2$
